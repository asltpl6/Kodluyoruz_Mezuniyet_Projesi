# Veri Seti Özelikleri

##### 1. Age (Yaş)
- Çalışanın yaşı (22 ile 60 arasında rastgele atanmış).

##### 2. Attrition (İşten Ayrılma)
- Çalışanın işten ayrılıp ayrılmadığını gösterir (Yes/No).

##### 3. Business Travel (İş Seyahati Sıklığı)
- Çalışanın iş seyahati durumu:
  - **Non-Travel**: Seyahat etmeyen.
  - **Travel_Rarely**: Seyrek seyahat eden.
  - **Travel_Frequently**: Sık seyahat eden.

##### 4. Department (Departman)
- Çalışanın çalıştığı departman:
  - **Software Development** (Yazılım Geliştirme)
  - **IT Support** (BT Destek)
  - **Project Management** (Proje Yönetimi)
  - **Data Science** (Veri Bilimi)
  - **Cybersecurity** (Siber Güvenlik)

##### 5. Education (Eğitim Seviyesi)
- Çalışanın eğitim seviyesi (1 = Lise, 2 = Üniversite, 3 = Lisansüstü, 4 = Doktora).

##### 6. Education Field (Eğitim Alanı)
- Çalışanın eğitim aldığı alan:
  - **Engineering** (Mühendislik)
  - **Computer Science** (Bilgisayar Bilimleri)
  - **Information Technology** (Bilgi Teknolojileri)
  - **Business Administration** (İşletme)
  - **Mathematics** (Matematik)

##### 7. Employee Number (Çalışan Kimlik Numarası)
- Çalışana atanan benzersiz kimlik numarası.

##### 8. Environment Satisfaction (İş Ortamı Memnuniyeti)
- Çalışanın iş ortamından memnuniyeti (1 ile 4 arasında).

##### 9. Gender (Cinsiyet)
- Çalışanın cinsiyeti (Male/Female).

##### 10. Job Involvement (İşe Katılım Düzeyi)
- Çalışanın işine ne kadar katılım sağladığını gösterir (1 ile 4 arasında).

##### 11. Job Role (İş Rolü)
- Çalışanın iş rolü, çalıştığı departman ve eğitim alanına bağlı olarak belirlenir. Örnek roller:
  - **Software Development**: "Software Engineer", "Backend Developer"
  - **IT Support**: "IT Specialist", "Helpdesk Analyst"
  - **Data Science**: "Data Scientist", "Machine Learning Engineer"
  - **Cybersecurity**: "Security Analyst"

##### 12. Job Satisfaction (İş Memnuniyeti)
- Çalışanın iş memnuniyetini gösterir (1 ile 4 arasında).

##### 13. Monthly Income (Aylık Maaş)
- Çalışanın aylık kazancı (3000 ile 15000 arasında rastgele atanmış).

##### 14. Num Companies Worked (Çalıştığı Şirket Sayısı)
- Çalışanın daha önce çalıştığı şirket sayısı (0 ile 6 arasında).

##### 15. OverTime (Fazla Mesai)
- Çalışanın fazla mesai yapıp yapmadığını gösterir (Yes/No).

##### 16. Percent Salary Hike (Yıllık Maaş Artışı Yüzdesi)
- Çalışanın maaşındaki yıllık yüzdelik artış (1 ile 20 arasında).

##### 17. Performance Rating (Performans Değerlendirme)
- Çalışanın performans değerlendirme puanı (1 ile 4 arasında).

##### 18. Relationship Satisfaction (İş İlişkileri Memnuniyeti)
- Çalışanın iş yerindeki ilişkilerden memnuniyeti (1 ile 4 arasında).

##### 19. Stock Option Level (Hisse Opsiyonu Seviyesi)
- Çalışanın hisse opsiyon seviyesi (0 ile 3 arasında).

##### 20. Total Working Years (Toplam Çalışma Yılları)
- Çalışanın toplam iş deneyim süresi (1 ile 40 arasında).

##### 21. Training Times Last Year (Geçen Yıl Alınan Eğitim Sayısı)
- Çalışanın geçen yıl aldığı eğitim sayısı (0 ile 5 arasında).

##### 22. Work-Life Balance (İş-Yaşam Dengesi Memnuniyeti)
- Çalışanın iş-yaşam dengesi memnuniyeti (1 ile 4 arasında).

##### 23. Years at Company (Şirketteki Yılları)
- Çalışanın şirkette toplam geçirdiği yıl (1 ile 20 arasında).

##### 24. Years in Current Role (Mevcut Roldeki Yılları)
- Çalışanın mevcut rolündeki yılı (1 ile 10 arasında).

##### 25. Years Since Last Promotion (Son Terfi Süresi)
- Çalışanın son terfisinden bu yana geçen yıl (0 ile 10 arasında).

##### 26. Years with Current Manager (Mevcut Yöneticiyle Geçirdiği Yıl)
- Çalışanın mevcut yöneticisiyle geçirdiği yıl (1 ile 10 arasında).

##### 27. Manager Gender (Yöneticinin Cinsiyeti)
- Çalışanın yöneticisinin cinsiyeti (Male/Female).

#### Kategorik, Numerik, ve Kardinal Değerlerin Yakalanması

```python
def grab_col_names(dataframe, cat_th= 8, car_th=15):
    # cat_cols,cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    print(f"Gözlemler: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')
    return cat_cols, num_cols, cat_but_car
cat_cols, num_cols, cat_but_car = grab_col_names(df)
```

```python
def cat_summary(dataframe, col_name):
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)}))
    print("##########################################")
    
for col in cat_cols:
    if df[col]. dtypes == "bool":
        df[col] = df[col].astype(int)
        cat_summary(df, col, plot=True)

    cat_summary(df, col)

    plt.figure(figsize=(10, 5))
    sns.countplot(x=col, data=df)
    plt.title(col)
    plt.show() 
```

### Modelin kurulması ve Performans verileri

#### Verinin Hazırlanması

*  Kategorik değişkenleri sayısal değerlere çevirme (Label Encoding)

```python
label_encoder = LabelEncoder()
df["GENDER"] = label_encoder.fit_transform(df["GENDER"])
df["MANAGER GENDER"] = label_encoder.fit_transform(df["MANAGER GENDER"])
df["DEPARTMENT"] = label_encoder.fit_transform(df["DEPARTMENT"])
df["JOB ROLE"] = label_encoder.fit_transform(df["JOB ROLE"])
df["OVER TIME"] = label_encoder.fit_transform(df["OVER TIME"])
```

#### Performans Değerlendirme Tahmini (Performance Rating Prediction)
Amaç: Çalışanların cinsiyeti, departmanları, yöneticilerinin cinsiyeti gibi faktörlere dayanarak performans değerlendirme puanlarını tahmin etmek. Cinsiyetin performans puanlarına etkisini incelemek.

```python
# Bağımsız ve bağımlı değişkenleri ayır
X_performance = df[["AGE", "GENDER", "MANAGER GENDER", "JOB SATISFACTION", "MONTHLY INCOME", "OVER TIME", "JOB INVOLVEMENT", "YEARS AT COMPANY"]]
y_performance = df["PERFORMANCE RATING"]

# Veriyi eğitim ve test setine ayır
X_train_perf, X_test_perf, y_train_perf, y_test_perf = train_test_split(X_performance, y_performance, test_size=0.2, random_state=42)
```

```python
# Modeli oluştur
rf_model_perf = RandomForestRegressor(random_state=42)

# Modeli eğit
rf_model_perf.fit(X_train_perf, y_train_perf)

# Test setinde tahmin yap
y_pred_rf_perf = rf_model_perf.predict(X_test_perf)

# Modeli değerlendirme
mse_rf_perf = mean_squared_error(y_test_perf, y_pred_rf_perf)
mae_rf_perf = mean_absolute_error(y_test_perf, y_pred_rf_perf)
rmse_rf_perf = np.sqrt(mse_rf_perf)
r2_rf_perf = r2_score(y_test_perf, y_pred_rf_perf)

print(f"Random Forest - Performance Rating Tahmini Sonuçları:")
print(f"Mean Squared Error (MSE): {mse_rf_perf}")
print(f"Mean Absolute Error (MAE): {mae_rf_perf}")
print(f"Root Mean Squared Error (RMSE): {rmse_rf_perf}")
print(f"R-squared (R²): {r2_rf_perf}")
```
