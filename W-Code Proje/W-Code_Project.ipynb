# Veri Seti Özelikleri

##### 1. Age (Yaş)
- Çalışanın yaşı (22 ile 60 arasında rastgele atanmış).

##### 2. Attrition (İşten Ayrılma)

##### 3. Business Travel (İş Seyahati Sıklığı)

  - **Travel_Rarely**: Seyrek seyahat eden.
  - **Travel_Frequently**: Sık seyahat eden.

##### 4. Department (Departman)
- Çalışanın çalıştığı departman:
  - **Software Development** (Yazılım Geliştirme)
  - **IT Support** (BT Destek)
  - **Project Management** (Proje Yönetimi)
  - **Data Science** (Veri Bilimi)
  - **Cybersecurity** (Siber Güvenlik)

##### 5. Education (Eğitim Seviyesi)
- Çalışanın eğitim seviyesi (1 = Lise, 2 = Üniversite, 3 = Lisansüstü, 4 = Doktora).

##### 6. Education Field (Eğitim Alanı)
- Çalışanın eğitim aldığı alan:
  - **Engineering** (Mühendislik)
  - **Computer Science** (Bilgisayar Bilimleri)
  - **Information Technology** (Bilgi Teknolojileri)
  - **Business Administration** (İşletme)
  - **Mathematics** (Matematik)

##### 7. Employee Number (Çalışan Kimlik Numarası)
- Çalışana atanan benzersiz kimlik numarası.

##### 8. Environment Satisfaction (İş Ortamı Memnuniyeti)
- Çalışanın iş ortamından memnuniyeti (1 ile 4 arasında).

##### 9. Gender (Cinsiyet)
- Çalışanın cinsiyeti (Male/Female).

##### 10. Job Involvement (İşe Katılım Düzeyi)
- Çalışanın işine ne kadar katılım sağladığını gösterir (1 ile 4 arasında).

##### 11. Job Role (İş Rolü)
- Çalışanın iş rolü, çalıştığı departman ve eğitim alanına bağlı olarak belirlenir. Örnek roller:
  - **Software Development**: "Software Engineer", "Backend Developer"
  - **IT Support**: "IT Specialist", "Helpdesk Analyst"
  - **Data Science**: "Data Scientist", "Machine Learning Engineer"
  - **Cybersecurity**: "Security Analyst"

##### 12. Job Satisfaction (İş Memnuniyeti)
- Çalışanın iş memnuniyetini gösterir (1 ile 4 arasında).

##### 13. Marital Status (Medeni Durum)
- Çalışanın medeni durumu (Single, Married, Divorced).

##### 14. Monthly Income (Aylık Maaş)
- Çalışanın aylık kazancı (3000 ile 15000 arasında rastgele atanmış).

##### 15. Num Companies Worked (Çalıştığı Şirket Sayısı)
- Çalışanın daha önce çalıştığı şirket sayısı (0 ile 6 arasında).

##### 16. OverTime (Fazla Mesai)
- Çalışanın fazla mesai yapıp yapmadığını gösterir (Yes/No).

##### 17. Percent Salary Hike (Yıllık Maaş Artışı Yüzdesi)
- Çalışanın maaşındaki yıllık yüzdelik artış (1 ile 20 arasında).

##### 18. Performance Rating (Performans Değerlendirme)
- Çalışanın performans değerlendirme puanı (1 ile 4 arasında).

##### 19. Relationship Satisfaction (İş İlişkileri Memnuniyeti)
- Çalışanın iş yerindeki ilişkilerden memnuniyeti (1 ile 4 arasında).

##### 20. Stock Option Level (Hisse Opsiyonu Seviyesi)
- Çalışanın hisse opsiyon seviyesi (0 ile 3 arasında).

##### 21. Total Working Years (Toplam Çalışma Yılları)

##### 22. Training Times Last Year (Geçen Yıl Alınan Eğitim Sayısı)
- Çalışanın geçen yıl aldığı eğitim sayısı (0 ile 5 arasında).

##### 23. Work-Life Balance (İş-Yaşam Dengesi Memnuniyeti)

##### 24. Years at Company (Şirketteki Yılları)

##### 25. Years in Current Role (Mevcut Roldeki Yılları)

##### 26. Years Since Last Promotion (Son Terfi Süresi)

##### 27. Years with Current Manager (Mevcut Yöneticiyle Geçirdiği Yıl)

```python
def grab_col_names(dataframe, cat_th= 8, car_th=10):
    """
    Docstring Açıklaması.
    Veri setindeki kategorik, numerik ve kategorik fakat kardinal değişkenlerin yakalanması.

    Parameters
    ------
        dataframe: dataframe
                Değişken isimleri alınmak istenilen dataframe
        cat_th: int, optional
                numerik fakat kategorik olan değişkenler için sınıf eşik değeri
        car_th: int, optinal
                kategorik fakat kardinal değişkenler için sınıf eşik değeri

    Returns
    ------
        cat_cols: list
                Kategorik değişken listesi
        num_cols: list
                Numerik değişken listesi
        cat_but_car: list
                Kategorik görünümlü kardinal değişken listesi
    """

    # cat_cols,cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')
    return cat_cols, num_cols, cat_but_car
cat_cols, num_cols, cat_but_car = grab_col_names(df)
```
